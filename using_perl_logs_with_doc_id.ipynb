```python
#family log
f = open('data/Obesity_data/ObesitySen_copy.dms','r')
content = f.read()
f.close()

#remove family history
records = content.strip().split('RECORD #')
f = open('data/Obesity_data/ObesitySen_remove_familiy_history.dms', "w") 
for record in records:
    id = record[:record.find(' ')]
    if(record.find('[report_end]') != -1):
        content = record[record.find(' ') + 1: record.find('[report_end]')].strip()
        sentences = content.split('\n')
        print(content[:20] + '\t' + id)
        f.write('RECORD #'+id+'\n')
        for sentence in sentences:
            if(sentence.find('FAMILY HISTORY:') != -1):
                continue
            f.write(sentence + '\n')
        f.write('[report_end]'+'\n')
f.close()   
```

```python
f = open('data/Obesity_data/ObesitySen_remove_familiy_history.dms', "r") 
content = f.read()
f.close()

#use only sentences with keyeords
records = content.strip().split('RECORD #')
f = open('data/Obesity_data/ObesitySen_only_q_n_p_intuitive_witd_doc_id.dms', "w")
for record in records:
    id = record[:record.find('\n')]
    if(record.find('[report_end]') != -1):
        content = record[record.find(' ') + 1: record.find('[report_end]')].strip()
        sentences = content.split('\n')
        print(content[:20] + '\t' + id)
        f.write('RECORD #'+id+'\n')
        for sentence in sentences:
            mark = 0
            # questionable log
            qfile = open('data/intuitive/questionable.log','r')
            content = qfile.read()
            lines = content.split('\n')
           
            for line in lines:
                if(line.find('|') != -1 and line.find(id.strip()) != -1):
                    print(line)
                    temp = line[line.find('|') +1 :].strip()
                    #print(temp)
                    keywords = temp.split('##')
                    for keyword in keywords:
                        if(keyword != '' and sentence.find(keyword) != -1):
                            print(keyword)
                            mark = 1
            # negated log
            nfile = open('data/intuitive/negated.log','r')
            content = nfile.read()
            lines = content.split('\n')
           
            for line in lines:
                if(line.find('|') != -1 and line.find(id + ' ')!=-1):
                    print(line)
                    temp = line[line.find('|') +1 :].strip()
                    #print(temp)
                    keywords = temp.split('##')
                    for keyword in keywords:
                        if(keyword != '' and sentence.find(keyword) != -1):
                            print(keyword)
                            mark = 1
            # positive log
            pfile = open('data/intuitive/positive.log','r')
            content = pfile.read()
            lines = content.split('\n')
           
            for line in lines:
                if(line.find('|') != -1 and  line.find(id + ' ')!=-1):
                    print(line)
                    temp = line[line.find('|') +1 :].strip()
                    #print(temp)
                    keywords = temp.split('##')
                    for keyword in keywords:
                        if(keyword != '' and sentence.find(keyword) != -1):
                            print(keyword)
                            mark = 1
            if(mark == 1):
                f.write(sentence + '\n')
        f.write('[report_end]'+'\n')
f.close()  
```
